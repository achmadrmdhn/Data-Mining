```python
# Mengimpor library python
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
# Membuat data dalam bentuk dictionary
data_manual = {
    'products': [
        'roti,susu,keju',
        'roti,susu',
        'susu,keju',
        'roti,keju',
        'roti,susu,keju,mentega',
        'roti,susu,mentega',
        'susu,keju,mentega',
        'roti,keju,mentega',
        'roti,susu,keju',
        'susu,mentega'
    ]
}

# Mengonversi data menjadi DataFrame
df = pd.DataFrame(data_manual)

# Menampilkan data
df
```

```python
# Menampilkan nilai DataFrame sebagai array
df.values
```

```python
# Memproses data menjadi sebuah list
data = list(df["products"].apply(lambda x: x.split(",")))

# Menmapilkan data
data
```

```python
# Mengubah data agar menjadi di analisis
te = TransactionEncoder()
data_encoded = te.fit(data).transform(data)
df_encoded = pd.DataFrame(data_encoded, columns=te.columns_)
df_encoded = df_encoded.astype(int)

# Menampilkan hasil
df_encoded
```

```python
# Mencari itemset yang sering muncul
frequent_itemsets = apriori(df_encoded, min_support=0.02, use_colnames=True)

# Menampilkan hasil
frequent_itemsets
```

```python
# Mencari association rule
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1.0)

# Menampilkan aturan asosiasi
rules
```
