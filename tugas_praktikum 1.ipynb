**1. Membaca dan Menampilkan Data**

```python
import pandas as pd

# Membaca data
df = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/Data/500_Person_Gender_Height_Weight_Index.csv")

# Mengganti nilai index
index_mapping = {
    0: 'Extremely Weak',
    1: 'Weak',
    2: 'Normal',
    3: 'Overweight',
    4: 'Obesity',
    5: 'Extreme Obesity'
}

# Mengubah kolom Index agar sesuai dengan mapping
df["Index"] = df["Index"].map(index_mapping)

# Menampilkan seluruh data
print("Data BMI (Body Max Index):")
print(df)
print("="*40)
```

**2. Menjalankan Fungsi Statistik**

```python
# Menjalankan fungsi statistik
# Menghitung jumlah data, rata-rata, nilai maksimum, nilai minimum
# Dari berat dan tinggi badan berdasarkan Gender

statistik = df.groupby('Gender').agg({
    'Height': ['count', 'mean', 'max', 'min'],
    'Weight': ['count', 'mean', 'max', 'min']
})

# Menampilkan fungsi statistik
print("Hasil Fungsi Statistik:")
print(statistik)
print("="*40)
```

**3. Mengubah Kolom Gender Menjadi Data Binary**

```python
# Mengubah kolom gender menjadi data binary

df['Gender'] = df['Gender'].map({'Female': 0, 'Male': 1})

# Menampilkan Data
print("Hasil Kolom Gender menjadi Data Binary:")
print(df)
print("="*40)
```

**4. Menampilkan Grafik**

```python
# Membuat grafik index BMI berdasarkan gender

import matplotlib.pyplot as plt
df.plot()

# Menampilkan grafik
plt.show()

print("Plotting:")
print("="*40)
```
